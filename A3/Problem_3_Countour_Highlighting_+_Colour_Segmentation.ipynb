```python
# Problem 3. Innovative Object Highlighting: Design a creative way to highlight the segmented objects
# within the image.

# USING CONTOUR HIGHLIGHTING

import cv2
import numpy as np
from google.colab.patches import cv2_imshow

image = cv2.imread('input_image.png')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
_, binary = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
contours, heirarchy = cv2.findContours(binary, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

highlighted_image = np.zeros_like(image)
cv2.drawContours(highlighted_image, contours, -1, (0, 255, 0), 2)

# input colour image
cv2_imshow(image)

# grayscale image
cv2_imshow(gray)

# binary image
cv2_imshow(binary)

# final contours image
cv2_imshow(highlighted_image)

```

```python
# Problem 3. Innovative Object Highlighting: Design a creative way to highlight the segmented objects
# within the image.

# USING COLOUR SEGMENTATION

import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Convert BGR to HSV Color Space
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Threshold of blue in HSV space
threshold_for_lower_blue = np.array([60, 35, 140])
threshold_for_upper_blue = np.array([180, 255, 255])

# Overlay Mask
mask = cv2.inRange(hsv, threshold_for_lower_blue, threshold_for_upper_blue)

result = cv2.bitwise_and(image, image, mask = mask)

cv2_imshow(image)
cv2_imshow(mask)
cv2_imshow(result)
```
